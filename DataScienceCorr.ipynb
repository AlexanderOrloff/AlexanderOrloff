WAB Yes/No Q не коррелирует ни с чем.

SpontSp Score for AQ (возможность к спонтанной речи) демонстрирует чёткую корреляцию на уровне 0.6 -0.75. Она корреллирует со всеми составляющими параметрами, причем корреляция тем выше, чем "сложнее" тест ( чем больше когнитивной нагрузки он требует)

Fluency показывает  более слабую корреляцию с лёгкими тестами (повторение слов, заканчивание предложений) и более высокую корреляция с более сложными тестами.

Sponteneous Fluency  почти ни с чем не коррелирует. Корреляция между восприятием на слух и порождением - самая слабая.

```python
Y = data_new['SpontSp Score for AQ']
X = data_new.copy()
X.drop('SpontSp Score for AQ', axis=1, inplace=True)

from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn import preprocessing
normalizer = preprocessing.MinMaxScaler()
X_real_norm_np = normalizer.fit_transform(X)
X = pd.DataFrame(data=X_real_norm_np, columns=X.columns)


lr = LinearRegression()
lr.fit(X,Y)
lr.coef_
print('R2: %f' % lr.score(X,Y))
```

```python
lr2 = Ridge(alpha=0.5, solver='sparse_cg')
lr2.fit(X,Y)
lr2.coef_
print('R2: %f' % lr2.score(X,Y))
```

```python
lr1 = Lasso(alpha=0.5)
lr1.fit(X,Y)
print('R2: %f' % lr1.score(X,Y))
```

```python
import numpy as np
pd.DataFrame(np.vstack((
        lr.coef_,
        lr1.coef_,
        lr2.coef_
        )), index=['LR', 'L1', 'L2'], columns=X.columns).plot(kind='bar', figsize=(20,10))
plt.title('Coefficient comparison', fontsize=20)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(\
                            X, Y, test_size=0.33, random_state=41)
lr = LogisticRegression()
lr.fit(X=X_train, y=y_train)
lr.score(X_test,y_test)
```

```python
from sklearn.cluster import KMeans
import numpy as np
datac = datac[[ 'WAB SpontSp Fluency', 'WAB       AudWdRec']]
inertia = []
for k in range(1, 7):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(datac)
    inertia.append(np.sqrt(kmeans.inertia_))

plt.plot(range(1, 7), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

```python
def iner(inertia):
    a = []
    d = {}
    minimum = 0
    for i in range(1, len(inertia)-1):
        d[i] = abs(inertia[i] - inertia[i+1])/abs(inertia[i-1] - inertia[i])
    minimum = min(d.values())
    for key in d.keys():
        if d[key] == minimum:
            return key + 1
            

iner(inertia)
```

```python
cl = KMeans(n_clusters=2, random_state=1234)

cl.fit(datac)

clusters = cl.fit_predict(datac)
cols = {1:'red', 0:'blue'}
color = [cols[x] for x in clusters]
datac.columns = ['x','y']
plt.figure(figsize=(10,10))
plt.scatter(x=datac['x'], y=datac['y'], c=color)
```
